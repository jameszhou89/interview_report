```python
import requests
import json
```

```python
def ths_gpt(temperature,max_tokens,prompt_text):
    url = "http://avatar.aicubes.cn/vtuber/auth/api/oauth/v1/login"

    payload = {
        "app_id": "your id here",
        "app_secret": "your secret here"
    }
    headers = {"content-type": "application/json"}

    response1 = requests.request("POST", url, json=payload, headers=headers)
    data1 = json.loads(response1.text)
    user_id = data1['data']['user_id']
    token = data1['data']['token']

    url = "http://avatar.aicubes.cn/vtuber/ai_access/chatgpt/v1/chat/completions"

    payload = {
        "messages": [
        {
            "role": "user",
            "content": prompt_text
        }
        ],
        "temperature": temperature,
        "max_tokens": max_tokens
    }

    headers = {
        "userId": user_id,
        "token": token,
        "content-type": "application/json"
    }

    response2 = requests.request("POST", url, json=payload, headers=headers)
    data2 = json.loads(response2.text)
    #print(data2)
    output = data2['data']['choices'][0]['message']['content']
    # print(output)
    return output
```
